# **A. Literature Review**
### Motivation
Global warming is a serious issue that is caused by Green House gas emission, including CO2. [1] The impact of global warming is enormous-temperature climbing up each year; iceberg melting, causing the sea level to rise and change not just climates but also species habitats, not to mention the impact on rising death rate on minority group such as children and elderly. [2] Some people argued that climate change is simply a myth and it's natural that the CO2 emission is increasing now. However, the amount of emission has been proofed to be caused by human activities and not by natural. [3] Human activities increase the greenhouse gas in the atmosphere, and then trap the heats on earth, causing the so-called "greenhouse effect", and result in the global-warming. [4] The importance of monitoring and controlling the emission of CO2 and other greenhouse gas thus speaks for itself. And that is why 195 countries signed on the famous Paris Agreement to reduce global greenhouse gas emissions. [5] 

After understanding the impact of global warming and its relation with CO2 emission, how are we going to predict the CO2 emission? Research shows that burning fuels, transportation, and consuming electricity can all contribute to the increase in CO2. [6] Thus, we would like to look into the CO2 emission in different industries to see how they correlated to the global CO2 emission.
Lastly, in 2020, researchers have suggested that the Covid-19 outbreak has lower the Co2 emission due to the overall lower human activities [7], and lower transportation [8]. Thus, we would also want to observe the impact of Covid-19 on CO2 emission in our 2020 data.

By monitoring the CO2 emission, we can build a model to help the government monitor the changes of global warming and greenhouse effect, and thus hopefully there will be one day, we can eliminate the human being's damage to the mother earth, really practically verify the efforts each country agreed to devote [9].
### Dataset
We first get the univariate dataset of global weekly Co2 emission from NOAA Global Monitoring Lab and combine it with the CO2 emission in different industries data from the US Energy Information Administration website to make it a multivariate dataset (monthly data with 555 data points). Then, we subset the univariate C02 emission data of 2020 and combine the subset with the COVID 19 2020 case number data by transfer the daily to weekly data (which will have 44 data points in this data).
### Our Approach
We will use Recurrent Neural Network models with layers such as LSTM and GRU to predict these two time-series datasets due to its strength in tracking down patterns [10].

### 4.Build a Baseline Model(DNN)

```python
# Now let's build a model without drop out
model = Sequential()
model.add(Dense(20, input_shape=(X.shape[1],), activation='relu')) # don't forget to add an input shape! (features,)
model.add(Dense(1, activation='linear'))
model.summary()

# compile the model
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])

# early stopping callback
from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss',
                   mode='min',
                   patience=50,
                   restore_best_weights = True)

# fit the model!
# remember, attach it to a new variable called 'history' in case
# we want to look at the learning curves

history = model.fit(X, y,
                    callbacks=[es],
                    epochs=5000,
                    batch_size=50,
                    validation_split=0.2,
                    verbose=1)
```

```python
pca = PCA(n_components=2) # the n_components number came from our above analysis
pca_result = pca.fit_transform(df_co2_emission_and_covid)
tmp = pd.DataFrame(columns={'pca-one', 'pca-two'})
tmp['pca-one'] = pca_result[:,0]
tmp['pca-two'] = pca_result[:,1] 
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
tmp.head()
```

```python
# prepare our X and Y variables. 
X = tmp
y = df_co2_emission_and_covid[['average co2 emission']]
print(df_co2_emission_and_covid.shape, X.shape, y.shape)
```

```python
# split a multivariate sequence into samples
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)
```

```python
# let's turn X into lookbacks of 10 with all of our samples
# all we need to do is decide on is n_steps (what our lookback period is)
# since we have a bunch of data, why not n_steps=10? then try 30 later on.
n_steps = 4
raw_seq = np.array(X) #make sure your data is stored as a numpy array!
# let's ignore the date column and just use the temperature data
X, y = split_sequences(raw_seq, n_steps)
```

```python
# note how there's a 2 at the end
# usually we did this for a multi-classification problem, but not today!
# by default, it's a 'linear' activiation function
# so this is 2 node output and we're doing regression.

n_steps = X.shape[1]
n_features = X.shape[2]
n_outputs = y.shape[1]

print(n_steps, n_features, n_outputs)
```

```python
# Build the model
n_steps = X.shape[1]
n_features = X.shape[2]
 

model = Sequential()
model.add(LSTM(50, activation='relu',
               recurrent_dropout = 0.1, 
               input_shape=(n_steps, n_features)))
model.add(Dropout(0.3))
# model.add(GRU(20, activation='relu'))
model.add(Flatten())
model.add(Dense(1)) 
# compile our model
opt = Adam(lr=1e-4)
model.compile(loss="mse", optimizer=opt)

import keras
from keras.callbacks import EarlyStopping

# early stopping callback
# This callback will stop the training when there is no improvement in  
# the validation loss for 10 consecutive epochs.  
es = keras.callbacks.EarlyStopping(monitor='val_loss', 
                                   mode='min',
                                   patience=50, # you can play with this!
                                   restore_best_weights=True) # important - otherwise you just return the last weigths...

# now we just update our model fit call
history = model.fit(X,
                    y,
                    callbacks=[es],
                    epochs=400, # you can set this to a big number!
                    batch_size=10,
                    validation_split=0.2,
                    verbose = 1)
```

### 3.Build the First Model(No Lags)

```python
del df_co2_emission['Month']
```

```python
Y = df_co2_emission[['Average CO2 Emissions']]
X = df_co2_emission.drop(axis=1, columns=['Average CO2 Emissions'])
print(df_co2_emission.shape, X.shape, Y.shape)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(X)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_)
```

```python
# let's see our variables in a 2D spaces.
X_new = pca.transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1],marker='o')
plt.show()
```

```python
pca = PCA(n_components=0.95)
pca.fit(X)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_)
print(pca.n_components_)
```

We let the components the explain 95% of the variance, the PCA gave us 3 n_components. We will use this number in the next step model preparation.

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(X)
tmp = pd.DataFrame(columns={'pca-one', 'pca-two', 'pca-three'})
tmp['pca-one'] = pca_result[:,0]
tmp['pca-two'] = pca_result[:,1] 
tmp['pca-three'] = pca_result[:,2]
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
tmp.head()
```

```python
tmp.shape
```

```python
X = tmp
X
```

```python
n_steps = 12
raw_seq = np.array(X) #data stored as numpy array!
X, y = split_sequences(raw_seq, n_steps)
```

```python
print(df_co2_emission.shape, X.shape, y.shape)
```

```python
n_steps = X.shape[1]
n_features = X.shape[2]
n_outputs = y.shape[1]

print(n_steps, n_features, n_outputs)
```

```python
# define model
model = Sequential()
model.add(Conv1D(filters=128, kernel_size=3, input_shape=(n_steps,n_features))) # notice how input shape goes in first layer
model.add(MaxPooling1D(2))
model.add(Bidirectional(LSTM(30,
                            return_sequences=True, # remember, if stacking layers, you need to return sequences!
                            activation='relu',
                            recurrent_dropout=0.2)))
model.add(GRU(20, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1)) 
model.add(Activation("linear"))
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
model.summary()

# early stopping callback
# This callback will stop the training when there is no improvement in  
# the validation loss for 10 consecutive epochs.  
es = keras.callbacks.EarlyStopping(monitor='val_loss', 
                                   mode='min',
                                   patience=10, 
                                   restore_best_weights=True) 

# now we just update our model fit call
history = model.fit(X,
                    y,
                    callbacks=[es],
                    epochs=800, 
                    batch_size=100,
                    validation_split=0.2,
                    verbose=1,
                    shuffle=True)
```

```python
# The training and validation lost are lower then previous models!
# scatterplots of actual vs. predicted
pred = model.predict(X)
pred
```

```python
y.shape
```

```python
pred.shape
```

```python
# show scatterplots of actual vs. predicted for train and test
# make a prediction

# calculate MSE for the baseline model
from sklearn.metrics import mean_absolute_error
print('MAE: ', mean_absolute_error(y, pred))

actual = y # the actual
plt.scatter(x=actual, y=pred)
x = np.linspace(-40,70) # 45 degree line from 0 to 25 (axes are the same)
plt.plot(x, x, color='red')
plt.suptitle('Acutal Value and Predicted Value of RNN')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
# looks pretty good!

# show timeseries plot on the actual and predicted data
plt.plot(np.arange(X.shape[0]), y, color='blue') # actual data
plt.plot(np.arange(X.shape[0]), pred, color='red') # predicted data
plt.suptitle('Show Timeseries Plot on the Acutal and Predicted data')
plt.xlabel('Time')
plt.ylabel('Average Co2 Emission')
plt.show()
```

* From the scatterplot and timeseries plot, we can see that this model is doing pretty good, with prediction aligning with actual CO2 emission.
