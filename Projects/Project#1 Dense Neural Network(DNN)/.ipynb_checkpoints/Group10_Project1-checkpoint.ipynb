# **A. Literature Review**
### Motivation

Nowadays, as social media platforms rise up as new ways to express personal emotions and connect with people, independent artists upload music to Youtube and other social media platforms to gain fame through "clicks". In 2020, the RAA reported that the music industry revenue grows in recorded, streaming, and wholesale even with the strike of COVID-19 ([link](https://www.riaa.com/wp-content/uploads/2020/09/Mid-Year-2020-RIAA-Revenue-Statistics.pdf)).[This article](https://econsultancy.com/the-impact-of-technology-and-social-media-on-the-music-industry/) explained that traditional consume of music changed rapidly as streaming platforms such as YouTube and Spotify started to prosper. The huge revenue comes with the viewing and the sharing on social media surely serves as a strong motivation for a lot of music artists out there all over the world since distance is no longer a barrier to spread their works.

It is intuitive that the popularity of a musician will affect his or her revenue. In 2020, the top 15 paid musicians earned more than 1 billion US dollars, with Kayne West listed as the number 1 paid musician with 170 million paid ([link](https://variety.com/2020/music/news/kanye-west-highest-paid-musician-celebrities-forbes-1234625689/)). The relation between the hotness of the songs and the revenue of the artists thus prove itself. Thus, predicting if the song is going to be a hit or not, might benefit the independent artists to be rewarded from their hardworkings. But how can we predict if the song is going to become a hit or not? How does a song's structure affect it's "click"? An [interesting finding](https://www.frontiersin.org/articles/10.3389/fnhum.2017.00263/full) from Neuroscience aspect suggests that the song structure changes, which they called "harmonic surprise", will increase the likelihood for a song to become a hit. [In this research](https://arxiv.org/pdf/1908.08609.pdf), the authors built up their dataset through Spotify and run several models to predict popularity, and get the best result from their random forest model. [On the other side](https://www.researchgate.net/publication/220723429_Hit_Song_Science_Is_Not_Yet_a_Science), however, it is stated that the information and features that used in music analysis are insufficient enough to grasp popularity of the songs. 


### Dataset
The creators of [the Million Song dataset](https://www.researchgate.net/publication/220723656_The_Million_Song_Dataset) introduced how they extract the data from the Echo Nest and the possible usage of this free dataset, including doing text analysis (artist name or lyrics),automatic music tagging, and year prediction...etc. Following by the introducing of this valuable source, researchers started to dig into this data.

 [A machine learning](http://www2.agroparistech.fr/ufr-info/membres/cornuejols/Teaching/Master-AIC/PROJETS-M2-AIC/PROJETS-2016-2017/Final-report-advanced%20machine%20learning_Xiaoxiao%20CHEN:Yuxiang%20WANG:Honglin%20LI.pdf) was conducted to help building recommendation system. However, the authors didn't get satisfying results from their models. [This research](https://towardsdatascience.com/song-popularity-predictor-1ef69735e380) concluded that it is easier to predict bad songs rather than good songs by fitting a XGBClassifier (An Application Programing Interface in Python)  to fit their model. [In this research](http://cs229.stanford.edu/proj2014/Angela%20Xue,%20Nick%20Dupoux,%20Predicting%20the%20Commercial%20Success%20of%20Songs%20Based%20on%20Lyrics%20and%20Other%20Metrics.pdf), the authors took a different approach to predict songs popularities by analyzing lyrics. However, the authors also stated that their model couldn't get predictive features either. Lastly, [this research](http://cs229.stanford.edu/proj2015/140_report.pdf) conducted a detailed analysis through different models, and concluded that features like artist familiarity, years, and loudness are not as predictive since there is a lot of variation of acoustic features within one songs. They also provide suggestion that by using a new metric of popularity such as number of downloads and number of litsens on music streaming platform might be helpful to build a more predictive models.

### Our Approach
Based on the literature reviews, we decided to use the Million Songs dataset to have a first look and understanding about building metrics to predict popularity, in this dataset, song hotness. Taking the experiences of these previous researches, we only analyze the song between 2006 to 2011 to avoid time trend effects, and we put the song hotness into three groups: if the song_hotttness is under 0.3, we consider it not popular and assign it 0; if the song_hotttness is between 0.3 and 0.7, we consider it okay and assign it 1; if the song_hotttness is above 0.7, we consider it a popular song and assign it 2. Moreover, we will fit this new assigned hotness value into Neural Network models to see if it correspond to the previous research. Lastly, though the finding suggest that "harmonic surprise" will increase the likelihood of making hit songs, we cannot observe such feature in the dataset.

## Deep Learning Model1

### Model Architecture

```python
#Input data
songs = pd.read_csv('/content/drive/Shared drives/OPIM 5509 Introduction to Deep Learning - Group10/Project 1/Datasets/finaldataset.csv')
songs.info()
```

```python
songsmodel=songs.drop(['artist_id', 'artist_latitude', 'artist_location','artist_longitude','artist_name','album_name','album_id','song_id',
                       'title'], axis=1, inplace=False)
songsmodel.info()
```

```python
# split into X and Y
Y = songsmodel['song_hotness']
X = songsmodel.drop(['song_hotness','song_hotness_num','artist_numid','year'], axis=1) #This time we also drop the year column

print(X.shape)
print(Y.shape)
```

```python
# convert to numpy arrays
X = np.array(X)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(Y)
dummy_y
```

```python
# Now let's build a model without drop out
model = Sequential()
model.add(Dense(20, input_shape=(X.shape[1],), activation='relu')) # don't forget to add an input shape! (features,)
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
# Again, compile the model
model.compile(optimizer='rmsprop', # use rmsprop or Adam
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
# This time minimizing the validation loss for 10 consecutive epochs.  
# This callback will stop the training when there is no improvement in  
# the validation loss for 10 consecutive epochs.
import keras
from keras.callbacks import EarlyStopping

es = keras.callbacks.EarlyStopping(monitor='val_loss', 
                                   mode='min',
                                   patience=10, 
                                   restore_best_weights=True)

# now we just update our model fit call
history = model.fit(X,
                    dummy_y,
                    callbacks=[es],
                    epochs=8000000, # Set a big number!
                    batch_size=10,
                    shuffle=True,
                    validation_split=0.2,
                    verbose=1)
```

### Evaluate the Model

```python
# learning curve
import matplotlib.pyplot as plt

# accuracy
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

# loss
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# "r" is for "solid red line"
plt.plot(epochs, acc, 'r', label='Training accuracy')
# b is for "solid blue line"
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
# Now the loss
plt.plot(epochs, loss, 'r', label='Training Loss')

plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

preds = model.predict(X)
#Print the accuracy score
print(accuracy_score(dummy_y.argmax(axis=1), preds.argmax(axis=1)))
#see the confusion matrix
matrix = confusion_matrix(dummy_y.argmax(axis=1), preds.argmax(axis=1))
matrix
```

```python
# more detail on how well things were predicted
print(classification_report(dummy_y.argmax(axis=1), preds.argmax(axis=1)))
```

### Model Architechture

## Song Title NLP
**Caution: this module need to take long time to run. Colab might fail to run the session.**

```python
import re
import nltk
```

```python
song_title = songs['title']
```

```python
list_rows = []
for index, row in song_title.iteritems():
    # split into words by white space
    words = row.split()
    
    # remove punctuation from each word
    import string
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in words]
    
    # convert to lower case
    stripped = [word.lower() for word in stripped]
    
    # Filter out stop words
    from nltk.corpus import stopwords
    stop_words = set(stopwords.words('english')) 
    removed_stop_words = [w for w in stripped if not w in stop_words]
    
    # remove remaining tokens that are not alphabetic
    words = [word for word in removed_stop_words if word.isalpha()]
    
    # Stemming of words
    from nltk.stem.porter import PorterStemmer
    porter = PorterStemmer()
    stemmed = [porter.stem(word) for word in words]
    list_rows.append(stemmed)
```

```python
songs['clean_title'] = list_rows
songs
```

### Topic Modeling - Latent Dirichlet Allocation (LDA)

```python
# Create a function to pull out nouns or adj from a string of text
from nltk import word_tokenize, pos_tag

def nouns_adj(text):
    '''Given a string of text, tokenize the text and pull out only the nouns and adjectives.'''
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)] 
    return ' '.join(nouns_adj)
```

```python
songs['clean_title']=songs['clean_title'].apply(str)
```

```python
# Apply the nouns_adj function to the transcripts to filter only on nouns
data_nouns_adj = songs['clean_title'].apply(nouns_adj)
data_nouns_adj
```

```python
# Import the necessary modules for LDA with gensim
# Terminal / Anaconda Navigator: conda install -c conda-forge gensim
from gensim import matutils, models
import scipy.sparse
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# Re-add the additional stop words since we are recreating the document-term matrix
add_stop_words = ['like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people',
                  'youre', 'got', 'gonna', 'time', 'think', 'yeah', 'said','new',
                  'say','says','said','first','second','will','year','years','now','one','may','just']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# Create a new document-term matrix using only nouns and adjectives, also remove common words with max_df
cvna = CountVectorizer(stop_words=stop_words, max_df=.8)
data_cvna = cvna.fit_transform(data_nouns_adj)
data_dtmna = pd.DataFrame(data_cvna.toarray(), columns=cvna.get_feature_names())
data_dtmna.index = data_nouns_adj.index
data_dtmna

# Create the gensim corpus
corpusna = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmna.transpose()))

# Create the vocabulary dictionary
id2wordna = dict((v, k) for k, v in cvna.vocabulary_.items())
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=10, id2word=id2wordna, passes=200)
ldana.print_topics()
```
